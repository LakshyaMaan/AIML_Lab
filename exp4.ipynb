```python
import pandas as pd

# Function to import data
def import_data(file_path):
    try:
        df = pd.read_csv(file_path)
        return df
    except Exception as e:
        print(f"Error importing file: {e}")
        return None

# Function to export data
def export_data(df, output_path):
    try:
        df.to_csv(output_path, index=False)
        print(f"Data exported successfully to {output_path}")
    except Exception as e:
        print(f"Error exporting file: {e}")

# Function to show dataset details
def show_dataset_details(df):
    if df is not None:
        print("Dataset Details:")
        print(f"Number of Rows: {df.shape[0]}")
        print(f"Number of Columns: {df.shape[1]}")
        print("\nFirst Five Rows:")
        print(df.head())
        print(f"\nDataset Size: {df.size}")
        print(f"\nNumber of Missing Values:\n{df.isnull().sum()}")
        print(f"\nSummary Statistics (Numerical Columns):\n{df.describe()}")
        print(f"\nSum of Numerical Columns:\n{df.sum(numeric_only=True)}")
        print(f"\nAverage of Numerical Columns:\n{df.mean(numeric_only=True)}")
        print(f"\nMinimum Values:\n{df.min(numeric_only=True)}")
        print(f"\nMaximum Values:\n{df.max(numeric_only=True)}")
    else:
        print("No data to show.")

# Example usage
file_path = 'database.csv'  # Replace with your input CSV file path
output_path = 'output_data.csv'

# Import dataset
df = import_data(file_path)

# Show dataset details
show_dataset_details(df)

# Export dataset
export_data(df, output_path)
```
