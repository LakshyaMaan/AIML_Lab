```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (make sure it's already uploaded or linked via Google Drive in Colab)
file_path = 'database.csv'  # Adjust if using Drive
df = pd.read_csv(file_path)

# Basic Structure of the Dataset
print("Dataset Overview:\n")
print(f"Number of Rows: {df.shape[0]}")
print(f"Number of Columns: {df.shape[1]}")
print("\nColumn Names and Data Types:\n")
print(df.dtypes)

# Checking for Missing Values
print("\nMissing Values:\n")
print(df.isnull().sum())

# Descriptive Statistics for Numerical Columns
print("\nSummary Statistics for Numerical Columns:\n")
print(df.describe())

# Check for Unique Values in Categorical Columns
print("\nUnique Values in Categorical Columns:\n")
for col in df.select_dtypes(include='object').columns:
    print(f"{col}: {df[col].unique()}")

# Handling Missing Values
# Option 1: Remove rows with missing values (simple approach)
df_dropped = df.dropna()
print(f"\nShape after dropping rows with missing values: {df_dropped.shape}")

# Option 2: Impute missing values with mean/median for numerical columns
df_imputed = df.fillna(df.mean(numeric_only=True))
print(f"\nShape after imputing missing values: {df_imputed.shape}")

# Visualizing the Distribution of Numerical Features
print("\nVisualizing Distributions:")
numerical_columns = df.select_dtypes(include=[np.number]).columns

# Plot histograms for numerical columns
df[numerical_columns].hist(figsize=(10, 8), bins=10, color='skyblue', edgecolor='black')
plt.tight_layout()
plt.show()

# Correlation Heatmap for Numerical Columns
plt.figure(figsize=(8, 6))
sns.heatmap(df[numerical_columns].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Boxplots to Visualize Outliers in Numerical Columns
plt.figure(figsize=(10, 6))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df, y=col, color='lightgreen')
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()

# Pairplot for Pairwise Relationships
sns.pairplot(df[numerical_columns].dropna())
plt.show()

# Categorical Data Analysis
print("\nCategorical Data Analysis:")
for col in df.select_dtypes(include='object').columns:
    print(f"\nFrequency count for {col}:\n")
    print(df[col].value_counts())

# Count Plot for Categorical Columns
plt.figure(figsize=(10, 6))
sns.countplot(x='City', data=df, palette='viridis')
plt.xticks(rotation=45)
plt.title('City Count Distribution')
plt.show()

```
